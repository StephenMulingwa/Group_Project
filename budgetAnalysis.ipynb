```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./addedroi.csv')
```

```python
data
```

```python
data['production_budget']
```

```python
budget = [float(value.replace("$", "").replace(",", "")) for value in data['production_budget']]
```

```python
vals = data['roi'].to_list()
```

```python
# Plotting
plt.figure(figsize=(12, 6))
plt.scatter(vals, budget)
plt.title('Average ROI by Budget')
plt.xlabel('roi')
plt.ylabel('Average ROI')
plt.xticks(rotation=45)
plt.grid(axis='x')
plt.show()
```
